```python
from codecoverage import CodeCoverage
from chromium import Chromium
from chrome_dev_protocol import ChromeDevToolsProtocol as CDP
from game_driver import GameDriver
```

```python
chromium = Chromium(headless=False)
cdp = CDP(chromium.url)
driver = GameDriver(cdp)
```

```python
driver.navigate("http://never-king.glitch.me")
```

```python
driver.add_sensor('codeCoverage',CodeCoverage())
```

```python
import time

seen = {}

while True:
    res = driver.step()["codeCoverage"]
    for script in res:
        for fun in script["functions"]:
            for r in fun["ranges"]:
                name = script["url"] + "#" + fun["functionName"] + "@" + str(r["startOffset"]) + "-" + str(r["endOffset"])
                if name not in seen:
                    print(name)
                    seen[name] = True
                    
    time.sleep(1)
```
